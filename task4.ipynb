## Prewitt filters

**<a id="#">Task 4.1.1.</a>** Load the image `data/astronaut.png` and show it:

**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img):
    result = np.zeros(img.shape)  
    img_prewitth_copy = img.copy()
    kernel_h = np.array ([[1,1,1], [0,0,0], [-1,-1,-1]])
    size = 3
    n = (size-1)//2
    for y in range (n, img.shape[0]-n):
        
        for x in range (n, img.shape[1]-n):
            neighborhood = np.zeros((size, size))
            
            for d in np.ndindex(size, size):
                dy = y + d[0] - n
                dx = x + d[1] - n
                neighborhood[d[0], d[1]] = img_prewitth_copy[dy, dx]
            
            prewitt_apply = np.sum (neighborhood * kernel_h)
            result[y,x] = prewitt_apply
    return result
def prewitt_v(img):
    result = np.zeros(img.shape)  
    img_prewittv_copy = img.copy()
    kernel_v = np.array ([[-1,0,1], [-1,0,1], [-1,0,1]])
    size = 3
    n = (size-1)//2
    for y in range (n, img.shape[0]-n):
        
        for x in range (n, img.shape[1]-n):
            neighborhood = np.zeros((size, size))
            
            for d in np.ndindex(size, size):
                dy = y + d[0] - n
                dx = x + d[1] - n
                neighborhood[d[0], d[1]] = img_prewittv_copy[dy, dx]
            
            prewitt_apply = np.sum (neighborhood * kernel_v)
            result[y,x] = prewitt_apply
    return result

# divide by 6? or no?
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
prewitth_img = prewitt_h(astr_ori)
prewittv_img = prewitt_v(astr_ori)
plt.figure(figsize = (15,8))
plt.subplot (1,2,1)
plt.imshow (prewitth_img)
```

**<a id="#">Task 4.1.4.</a>** Compare your results with the correct result image `data/astronaut_prewitt_h.tiff` and `data/astronaut_prewitt_v.tiff`:

```python
ideal_prewitt_h = skimage.io.imread('data/astronaut_prewitt_h.tiff')
ideal_prewitt_v = skimage.io.imread('data/astronaut_prewitt_v.tiff')

plt.figure(figsize=(15,8))

plt.subplot(2,2,1)
plt.title("Reference h")
plt.imshow(ideal_prewitt_h)

plt.subplot(2,2,2)
plt.title("Own h")
plt.imshow(prewitth_img)

plt.subplot(2,2,3)
plt.title("Reference v")
plt.imshow(ideal_prewitt_v)

plt.subplot(2,2,4)
plt.title("Own v")
plt.imshow(prewittv_img)
```

## Sobel filters (bonus)

**<a id="#">Task 4.3.1.</a>** Implement the two functions `sobel_h` and `sobel_v` for horizontal and vertical 3&times;3 Sobel filtering, respectively:

```python
def sobel_h(img):
    result = np.zeros(img.shape)  
    img_sobelh_copy = img.copy()
    kernel_h = np.array ([[1,2,1], [0,0,0], [-1,-2,-1]])
    size = 3
    n = (size-1)//2
    for y in range (n, img.shape[0]-n):
        
        for x in range (n, img.shape[1]-n):
            neighborhood = np.zeros((size, size))
            
            for d in np.ndindex(size, size):
                dy = y + d[0] - n
                dx = x + d[1] - n
                neighborhood[d[0], d[1]] = img_sobelh_copy[dy, dx]
            
            sobel_apply = np.sum (neighborhood * kernel_h)
            result[y,x] = sobel_apply
    return result
def sobel_v(img):
    result = np.zeros(img.shape)  
    img_sobelv_copy = img.copy()
    kernel_v = np.array ([[-1,0,1], [-2,0,2], [-1,0,1]])
    size = 3
    n = (size-1)//2
    for y in range (n, img.shape[0]-n):
        
        for x in range (n, img.shape[1]-n):
            neighborhood = np.zeros((size, size))
            
            for d in np.ndindex(size, size):
                dy = y + d[0] - n
                dx = x + d[1] - n
                neighborhood[d[0], d[1]] = img_sobelv_copy[dy, dx]
            
            sobel_apply = np.sum (neighborhood * kernel_v)
            result[y,x] = sobel_apply
    return result
```

**<a id="#">Task 4.3.2.</a>** Test your above solution using the functions `sobel_h` and `sobel_v` for the previously loaded image and show the result:

```python
sobelh_img = sobel_h(astr_ori)
sobelv_img = sobel_v(astr_ori)
plt.figure(figsize=(15,8))
plt.subplot(1,2,1)
plt.imshow(sobelh_img)
plt.subplot(1,2,2)
plt.imshow(sobelv_img)
```
